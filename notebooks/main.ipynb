## DEM - Elastic properties (Figures in section 3)

First, the results must be unzip. Then, the results can be plotted with plot_elastic_properties.py.

```python
d0 = 0.003
L = [2*d0, round(3*d0,3), round(6*d0,3), 9*d0, round(12*d0,3), 15*d0, 18*d0, round(21*d0,3), round(24*d0,3), 27*d0]

for i in range(len(L)):
    with zipfile.ZipFile('../data/elastic_properties/box_' + str(L[i]) + '.zip', 'r') as zip_ref:
        print('../data/elastic_properties/box_' + str(L[i]) + '.zip')
        zip_ref.extractall('../data/elastic_properties/')
        print('Done')
```
